### YOLOv11 - PPE

```python
from ultralytics import YOLO
```

```python
yaml_path = "dataset/data.yaml"

model = YOLO("yolo11n.pt")

print(f"Iniciando treinamento > {yaml_path}")

results = model.train (
    data = yaml_path,

    device='cpu',      
    workers=0,         
    batch=4,           
    
    epochs=100,         
    imgsz=640,         
    patience=10,        
    
    project='runs/detect',       
    name='treino_epi_v1',        
    exist_ok=True,               
    verbose=True


)
```
